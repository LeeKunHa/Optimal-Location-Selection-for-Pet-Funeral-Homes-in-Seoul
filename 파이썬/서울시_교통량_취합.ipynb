```python
import pandas as pd
import numpy as np
import os

# 파일들이 있는 폴더명으로 폴더내 파일 목록 확인
forders = os.listdir('서울시 교통량')
print(forders)
```

```python
df_all = pd.DataFrame()
for i in range(0,len(forders)):
    if forders[i].split('.')[1] == 'csv':
        file = '서울시 교통량/'+forders[i]
        df= pd.read_csv(file,encoding='cp949') 
        df_all = pd.concat([df_all, df])
```

```python
df_all.to_csv("서울시_교통량.csv",encoding='cp949', index = False)
```
